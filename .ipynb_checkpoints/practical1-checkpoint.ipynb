```python
import numpy as np
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
import pandas as pd
```

```python
data = {
    'Name': ['A', 'B', 'C', 'D'],
    'Age': [20, 21, None, 22],
    'Gender': ['Male', 'Female', 'Female', 'Male'],
    'Marks': [85, 90, 88, None]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isnull()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.shape
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
```

```python
df['Age'] = df['Age'].astype(int)
```

```python
le = LabelEncoder()
```

```python
df['Gender'] = le.fit_transform(df['Gender'])
```
