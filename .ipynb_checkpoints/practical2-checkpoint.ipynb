```python
import pandas as pd
```

```python
import numpy as np
```

```python
data = {
    'Name':['A', 'B', 'C', 'D'],
    'Marks':[23, None, 32, 84],
    'StudyHours' :[3, 2, 5, None]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
```

```python
df['StudyHours'] = df['StudyHours'].fillna(df['StudyHours'].mean())
```

```python
q1 = df['Marks'].quantile(0.25)
```

```python
q3 = df['Marks'].quantile(0.75)
```

```python
iqr = q3 - q1
```

```python
lower = q1 - 1.5 * iqr
```

```python
upper = q3 + 1.5 * iqr
```

```python
df = df[(df['Marks'] >= lower) & (df['Marks'] <= upper)]
```

```python
df['StudyHours'] = np.log(df['StudyHours'])
```
