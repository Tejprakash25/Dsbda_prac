```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("iris.csv")

print(df.head())

print(df.dtypes)
```

```python
features = [
    'SepalLengthCm',
    'SepalWidthCm',
    'PetalLengthCm',
    'PetalWidthCm'
]
```

```python
for i in features:

    plt.hist(df[i])

    plt.title(i)

    plt.show()
```

```python
for i in features:

    plt.boxplot(df[i])

    plt.title(i)

    plt.show()
```
