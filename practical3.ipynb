```python
import pandas as pd
```

```python
data = {
    'Name': ['A','B','C','D'],
    'Department': ['IT','IT','COMP','COMP'],
    'Income': [25000,30000,20000,28000],
    'Experience': [2,5,3,6]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('Department')['Income'].agg(
    ['mean','median','min','max','std']
)
```

```python
for d, x in df.groupby('Department'):
    print(d, x['Income'].tolist())
```

```python
df.groupby('Department')['Experience'].describe()
```
