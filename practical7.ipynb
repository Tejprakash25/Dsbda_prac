```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
```

```python
text = "Data Science is very useful for data analysis and machine learning."

tokens = word_tokenize(text)

print("Tokens")
print(tokens)

pos = pos_tag(tokens)

print("POS Tagging")
print(pos)
```

```python
stop_words = stopwords.words('english')

filtered = [w for w in tokens if w.lower() not in stop_words]

print("Stop Words Removed")
print(filtered)
```

```python
ps = PorterStemmer()

stemmed = [ps.stem(w) for w in filtered]

print("Stemming")
print(stemmed)
```

```python
lm = WordNetLemmatizer()

lemmatized = [lm.lemmatize(w) for w in filtered]

print("Lemmatization")
print(lemmatized)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "Data Science is useful",
    "Machine learning uses data",
    "Data analysis is important"
]

tfidf = TfidfVectorizer()

result = tfidf.fit_transform(documents)
```

```python
print("TF-IDF")
print(result.toarray())

print("Feature Names")
print(tfidf.get_feature_names_out())
```
