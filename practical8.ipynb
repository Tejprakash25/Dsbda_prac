```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("titanic")

print(df.head())

print(df.info())
```

```python
sns.countplot(x='sex', data=df)

plt.show()

sns.countplot(x='class', data=df)

plt.show()

sns.histplot(df['fare'], bins=20)

plt.show()
```
